# How Economic Development is impacting Biodiversity
### Adressing the impact of Agricultural land expansion on Forest cover

Biodiversity, the "lungs of the Earth", describes a complex system on which humans have an impact. A complex dynamic is unfolding between the expansion of agricultural land and the sustainability of our forests. With this analysis, we explore the key factors behind this relationship, highlighting the urgent need to reconsider our habits. The results of our research reveal a worrying fact: every advance in economic development seems to leave a negative imprint on our planet's forest cover. What's more, they suggest that current policies do not appear to be a sufficient solution for combining population growth and environmental preservation. Together, let's delve into the complexity of this relationship, where every tree felled has far-reaching consequences, prompting us to think deeply about the implications of our system of pursuing economic development at all costs.


## 1. Introduction

Biodiversity englobes all life forms including animals, plants, fungi, and bacteria, collectively composing the natural world. These diverse species interact within ecosystems, playing critical roles in sustaining life <a href="https://www.worldwildlife.org/pages/what-is-biodiversity#:~:text=Biodiversity%20is%20all%20the%20different,maintain%20balance%20and%20support%20life" target="_blank">[1]</a>. Biodiversity includes pollinators responsible for crops, bushes reducing flooding, trees absorbing CO2, or coral reefs protecting coastlines from waves <a href="https://royalsociety.org/topics-policy/projects/biodiversity/why-is-biodiversity-important/" target="_blank">[2]</a>. Its attributes, such as habitat area and species abundance, are crucial in delivering ecosystem services such as provisioning (e.g.: timber production), regulating (e.g.: water purification) or cultural (e.g.: recreation) services (Harrison et al., 2014) <a href="https://doi.org/10.1016/j.ecoser.2014.05.006" target="_blank">[3]</a>. Humans rely on healthy ecosystems for their survival as they provide clean air, fresh water, medicines and food security <a href="https://www.who.int/news-room/fact-sheets/detail/biodiversity-and-health#:~:text=Healthy%20communities%20rely%20on%20well,disease%20and%20stabilize%20the%20climate." target="_blank">[4]</a>.

Although species have always faced natural extinction, the current rate is unprecedented, and the proportion of species threatened to disappear is increasing <a href="https://royalsociety.org/topics-policy/projects/biodiversity/what-is-the-scale-of-biodiversity-loss/" target="_blank">[5]</a>.  The WWF’s 2022 Living Planet Report <a href="https://livingplanet.panda.org/en-US/">[6]</a> notes a 69% decline in mammals, fish, birds, reptiles, and since 1970. Additionally, 75% of freshwater resources and 33% of land surface are used for crop and livestock production further stressing biodiversity. Species loss threatens the overall functioning of ecosystems (Gamfeldt et al., 2008) <a href="http://www.jstor.org/stable/27651669" target="_blank">[7]</a>, and the services they provide, which impacts the benefits humans derive from them (Díaz et al., 2006) <a href="https://doi.org/10.1371/journal.pbio.0040277" target="_blank">[8]</a>.


> "*Estimates suggest that extinction rates in the recent past have been running tens to hundreds of times faster than in pre-human times and that the pace is accelerating.*" <a href="https://royalsociety.org/topics-policy/projects/biodiversity/what-is-the-scale-of-biodiversity-loss/" target="_blank">[9]</a>

The main drivers of biodiversity loss include human activities like deforestation, pesticide use, agriculture, over-hunting, spread of invasive species, and resource extraction. Land use change and overexploitation together account for 50% of global biodiversity decline, while climate change and pollution contribute another 14% <a href="https://royalsociety.org/topics-policy/projects/biodiversity/human-impact-on-biodiversity/" target="_blank">[10]</a>. Habitat alteration and the increased frequency of extreme events (e.g.: wildfire, flooding) caused by climate change, will add even more pressure on diminishing populations <a href="https://royalsociety.org/topics-policy/projects/biodiversity/is-the-rate-of-biodiversity-loss-increasing-or-decreasing/" target="_blank">[11]</a>.

With the growing human population and its demand for resources, more natural habitats are devoted to agriculture, mining, industrial infrastructure, and urban areas <a href="hhttps://royalsociety.org/topics-policy/projects/biodiversity/human-impact-on-biodiversity/" target="_blank">[12]</a>, exacerbating biodiversity degradation. Furthermore, economic development leads to increased consumption and higher demands on Earth’s biological productivity <a href="https://royalsociety.org/topics-policy/projects/biodiversity/how-does-the-growing-global-population-and-increasing-consumption-affect-biodiversity/" target="_blank">[13]</a>.
Given the direct implications of biodiversity loss on food security and human health, we thought it was relevant to conduct research on the impact of economic development on biodiversity. However, economic development includes many aspects that impact biodiversity in different ways. We decided to focus on one specific relationship: the impact of agricultural land expansion on forest cover. Agriculture is closely tied to economic development, as indicated by the World Bank Data <a href="https://www.worldbank.org/en/topic/agriculture" target="_blank">[14]</a>, contributing to GDP, employment, infrastructure and other factors. Simultaneously, forests are vital for biodiversity, offering food, shelter, clean air and water to a wide range of species (Brockerhoff et al., 2017) <a href="https://doi.org/10.1007/s10531-017-1453-2" target="_blank">[15]</a>. A decline in forest cover directly contributes to biodiversity loss. Forest diversity is essential for ecosystem regulation (Ehbrecht et al., 2021) <a href="https://doi.org/10.1038/s41467-020-20767-z" target="_blank">[16]</a>
, and agriculture is key for poverty reduction and food security <a href="https://www.worldbank.org/en/topic/agriculture" target="_blank">[17]</a>. With the expected increase in global food demand (Gomiero et al., 2011) <a href="https://doi.org/10.1080/07352689.2011.553515" target="_blank">[18]</a>, the challenge is to augment agricultural production while minimizing deforestation.

Our hypothesis is that agricultural land expansion leads to a decrease in forest cover negatively impacting biodiversity. This has been previously established in other studies. With our research we contribute to this area of study by confirming the findings with updated data and by estimating the impact of a reforestation policy in Ghana, a country particularly affected by deforestation due to agriculture.

## 3. Linear regression model

### 3.1 Methodology of the linear regression model

Our analysis began with the following linear regression model (Equation 1):

<br>

\begin{equation}
Forest Cover_{it} = \beta_0 + \beta_1 Agricultural Land_{it} + \beta_2 \log(Population Density_{it}) + \beta_3 GDP per capita_{it} + \beta_4 Latitude_i + \beta_5 \log(Fire_{it}) + \alpha_i + \delta_t + u_{it}
\end{equation}

<br>

Where $Forest Cover_{it}$ is our dependent variable, denoting the share of forest in country $i$ (% of the country's total land area) in year $t$. $Agricultural Land_{it}$ is our main independent variable, designating the share of land dedicated to agriculture in country $i$ (% of the country's total land area), for year $t$. The main effect we seek to identify with this model is $\beta_1$, i.e. the impact of the variation of a country's agricultural land on its forests. The addition of other factors that could influence this relationship as control variables is therefore essential in order to isolate the desired effect. The $Population Density_{it}$ variable reflects the population density (inhabitants/km²) of a country $i$ in year $t$. We consider a country's population density to be significant, because the higher it is, the more likely it is that the country will destroy forests for urban development, as is the case for Singapore and Monaco <a href="https://worldpopulationreview.com/country-rankings/countries-by-density" target="_blank">[19]</a>. The $Latitude_i$ variable considers a country's distance from the equator (degrees), influencing its climate and, consequently, its agriculture and forests (see Figure 2).

**Figure 2**: Map of forest cover (in green) <a href="https://www.treehugger.com/maps-of-the-worlds-forests-1343036" target="_blank">[20]</a>

<img src = "https://raw.githubusercontent.com/G21clau5/Project_Whale/main/Map%20of%20forest%20cover.webp" width="600">


$GDP per capita_{it}$ (US$/capita) takes into account the fact that rich countries are less inclined to expand their agricultural land, preferring instead to import their food. As Zabel et al. (2019) <a href="https://doi.org/10.1038/s41467-019-10775-z" target="_blank">[21]</a> point out (see Figure 3), developing countries, such as sub-Saharan Africa, South America and South-East Asia, tend to have a high level of biodiversity and to expand their agricultural land to meet the needs of their growing populations. Wealthy countries, such as those in Western Europe and North America, on the other hand, have less biodiversity and rely more on imported food.

**Figure 3**: Spatial association between endemism richness and potentials for cropland expansion and intensification <a href="https://doi.org/10.1038/s41467-019-10775-z" target="_blank">[22]</a>

<img src = "https://raw.githubusercontent.com/G21clau5/Project_Whale/main/Endemism%20richness%20-%20expansion%20potential.webp" width="600">

High–low and low–high clusters show inverse spatial association. Three shades of colors indicate significant results for one, two or all three taxonomic groups (birds, mammals, amphibians).

Finally, $Fire_{it}$ takes into account "natural" forest fires, which are not directly linked to agriculture but which nevertheless have a significant impact on deforestation. Country and year fixed effects are also included to control for the effects of unobserved variables, such as economic conditions or culture vis-à-vis agriculture.



### 3.2 Results of the linear regression model

Without control variables or fixed effects, our model estimates a negative relationship (see Graph 1), significant at the 5% threshold (see Appendix 2). This implies that a 1% increase in agricultural land leads to a 0.68% decrease in forest cover. Despite the fact that the percentage of agricultural land alone explains 46.1% of the variation in forest cover, it is essential to consider other factors, such as the control variables and the country and year fixed effects.

**Graph 1**: Linear regression of the percentage of forest cover on the percentage of agricultural land.

```python
#Scatterplot of FC on AL with global regression
fig, ax = plt.subplots(figsize=(10, 5))     #largeur, hauteur
sns.scatterplot(x='Agricultural Land', y='Forest Cover', legend=False, data=dfi, s=8, color='black')
sns.regplot(x='Agricultural Land', y='Forest Cover', data=dfi, scatter=False, color='blue', label='Régression linéaire, IC95%', ci=95) # Ajouter une régression linéaire

import scipy.stats
# Correlation coefficient (Pearson)
correlation_coefficient = dfi['Agricultural Land'].corr(dfi['Forest Cover'])
plt.text(0.7, 0.91, f'Correlation : {correlation_coefficient:.2f}', fontsize=12, transform=plt.gca().transAxes)

plt.legend(loc='lower right')
text_below = "Notes: This graph shows the relationship between the Forest Cover and % of agricultural land. \nThe blue line is a linear OLS regression with an IC of 95% and the correlation coefficient of these two variables. \nThe source of the data is World Bank Data. The sample contains 133 countries for the period 2001-2021. "
ax.text(0.5, -0.2, text_below, horizontalalignment='center', verticalalignment='center', transform=ax.transAxes, fontsize=12)
plt.xlabel('Agricultural Land [%]')
plt.ylabel('Forest Cover [%]')
plt.title('Forest Cover vs Agricultural Land [%]')
plt.show()
```

When we look at the full model (Equation 1) (see Appendix 3), although the relationship between agricultural land and forest cover remains significant, it flattens out, implying a 0.07% decrease in forest cover for a 1% increase in agricultural land, thus not rejecting our hypothesis. All the control variables, with the exception of forest fires, are significant. This could be due to the high variability of the variable over time (see Appendix 4).


This model enabled us to estimate a negative effect of agricultural land use on forest extent. However, this regression has certain limitations, particularly in terms of causality (see Section 6). For this reason, we decided to take our analysis a step further. We examine the effect of an environmental policy on this dynamic using the difference-in-differences method, as an extension of our research.


## 4. Difference-in-difference

### 4.1 Methodology of the Difference-in-difference

The study by Curtis et al. (2018) <a href="https://doi.org/10.1126/science.aau3445" target="_blank">[23]</a>, using satellite data to show the evolution of forests, reveals significant deforestation in Sub-Saharan Africa (see Figure 4), mainly due to "shifting cultivation". This method, which involves clearing land (sometimes by burning), over-exploiting it and then abandoning it, is seriously damaging to forests and biodiversity. This notable relationship observed in Sub-Saharan African countries (see Graph 2) corroborates this study and the observations of Zabel et al. (2019) <a href="https://doi.org/10.1038/s41467-019-10775-z" target="_blank">[24]</a>.

**Figure 4**: Primary drivers of forest cover loss for the period 2001 to 2015 <a href="https://doi.org/10.1126/science.aau3445" target="_blank">[25]</a>

<img src = "https://raw.githubusercontent.com/G21clau5/Project_Whale/main/Primariy%20drivers%20of%20forest%20cover%20loss.jpeg" width="600">

Note: Darker color intensity indicates greater total quantity of forest cover loss.

**Graph 2**: Bivariate relationship of sub-Saharan countries compared with other countries, for countries with a relationship more negative than the median.

```python
# scatterplot for sub-saharien coutnires vs others (only countries with slope more negative than the median)

# Slope for each "Iso"
slopes = []

for iso in dfi['Iso'].unique():
    data_iso = dfi[dfi['Iso'] == iso]
    if len(data_iso) > 1:
        slope, intercept = np.polyfit(data_iso['Agricultural Land'], data_iso['Forest Cover'], 1)
        slopes.append((iso, slope))

# Median of slopes
median_slope = np.median([slope for iso, slope in slopes])
filtered_df = dfi[dfi['Iso'].map(lambda x: np.polyfit(dfi[dfi['Iso'] == x]['Agricultural Land'], dfi[dfi['Iso'] == x]['Forest Cover'], deg=1)[0] <= median_slope)]
filtered_df['Color'] = filtered_df['Iso'].apply(lambda x: 'red' if x in pays_sub_agricultural else 'grey')

# Scatterplot
fig, ax = plt.subplots(figsize=(10, 7))
sns.scatterplot(x='Agricultural Land', y='Forest Cover', legend=False, hue='Color', palette={'red': 'red', 'grey': 'grey'}, data=filtered_df)
plt.xlabel('Agricultural Land')
plt.ylabel('Forest Cover')
plt.title('Forest Cover vs % of agricultural land')

# arrows from 2001 to 2021
for iso in filtered_df['Iso'].unique():
    data_2001 = dfi[(dfi['Iso'] == iso) & (dfi['Year'] == 2001)]
    data_2021 = dfi[(dfi['Iso'] == iso) & (dfi['Year'] == 2021)]

    if not data_2001.empty and not data_2021.empty:
        x_2001, y_2001 = data_2001['Agricultural Land'].values[0], data_2001['Forest Cover'].values[0]
        x_2021, y_2021 = data_2021['Agricultural Land'].values[0], data_2021['Forest Cover'].values[0]

        # arrows options
        x_offset = -0.01  # Horizontal
        y_offset = -0.005  # Vertical
        plt.annotate('', xy=(x_2021 + x_offset, y_2021 + y_offset), xytext=(x_2001 + x_offset, y_2001 + y_offset),
                     arrowprops=dict(arrowstyle='->', color='gray'))

# Labels for each country
for iso, data in dfi.groupby('Iso'):
    last_row = data.tail(1)
    label = f'  {iso}'
    ax.annotate(label, (last_row['Agricultural Land'], last_row['Forest Cover']), fontsize=8)

plt.legend()
text_below = "Notes: This graph shows the relationship between the Forest Cover and % of agricultural land. Red dots represent sub-saharan countries. \nThe source of the data is World Bank Data. The sample contains 133 countries for the period 2001-2021."
ax.text(0.5, -0.3, text_below, horizontalalignment='center', verticalalignment='center', transform=ax.transAxes, fontsize=12)

plt.show()
```

Within this region, three countries - Ghana, Rwanda and Kenya - have managed to reverse their deforestation trends (see Graph 3). Ghana is particularly notable for its rapid deforestation before 2010.

**Graph 3**: Evolution of forest cover (%) for countries where the trend has been reversed.

```python
pays_sub_agricultural = ['AGO','BEN','BDI','CMR','CAF','COD','COG','CIV','SWZ','ETH','GAB',
            'GHA','GIN','GNB','KEN','MWI','MOZ','NGA',
            'RWA','SLE','ZAF','TZA','TGO','UGA','ZMB','ZWE']

def has_trend_inversion(country_data):
    differences = country_data['Forest Cover'].diff().dropna()
    return any(differences.iloc[i] * differences.iloc[i + 1] < 0 for i in range(len(differences) - 1))

# List with countries with an inversion
countries_with_inversion = []

for iso in pays_sub_agricultural:
    country_data = dfi[dfi['Iso'] == iso]
    if has_trend_inversion(country_data):
        countries_with_inversion.append(iso)

fig = plt.figure(figsize=(10, 7))

for iso in countries_with_inversion:
    country_data = dfi[dfi['Iso'] == iso]
    sns.lineplot(x='Year', y='Forest Cover', data=country_data)

    # Country labels
    if country_data[country_data['Year'] == 2021].shape[0] > 0:
        last_value = country_data[country_data['Year'] == 2021]['Forest Cover'].values[0]
        country_name = country_data[country_data['Year'] == 2021]['Country'].values[0]
        plt.text(2021 + 0.1, last_value, country_name, fontsize=9, va='center')

# Utilisation de la variable fig pour ajouter du texte en dessous du graphique
plt.title('Evolution of Forest Cover in Countries with Trend Inversions (Yearly)', fontsize=16)
plt.xlabel('Year', fontsize=14)
plt.ylabel('Forest Cover [%]', fontsize=14)
plt.grid(True)

# Utilisation de fig.text au lieu de ax.text
text_below = "Notes: This graph shows the evolution of the Forest Cover for Sub-saharian countries with an inversion of the trend.\nThe source of the data is World Bank Data. The sample contains 133 countries for the period 2001-2021. "
fig.text(0.5, -0.0, text_below, horizontalalignment='center', verticalalignment='center', transform=fig.transFigure, fontsize=12)

# Affichage du graphique
plt.show()
```

Formally, a direct comparison would indicate the following result (Equation 2):

<br>

\begin{equation}
E[Y_{it} | \text{Treated}_i, \text{Post}_t] - E[Y_{it} | \text{Treated}_i, \text{Pre}_t] = \lambda_t + \beta
\end{equation}

<br>

Where $\beta$ is the effect of the reform we seek to identify and $\lambda_t$ are effects biasing the results, such as "bad" seasons affecting agricultural land, storms destroying forests or any other undesirable effects.

The use of a control group makes up for this. Côte d'Ivoire, Ghana's neighbouring country with many similar characteristics such as population (33 million in Ghana compared with 28 million in Côte d'Ivoire), GDP (73 billion in Ghana, 70 billion in Côte d'Ivoire) and changes in forest cover (see Graph 4), thus validating the parallel trend hypothesis (see Appendix 5), is a potential candidate for this role.

Formally, the difference for the control group (Côte d'Ivoire) is as follows (Equation 3):

<br>

\begin{equation}
E[Y_{it} | \text{Control}_i, \text{Post}_t] - E[Y_{it} | \text{Control}_i, \text{Pre}_t] = \lambda_t
\end{equation}

<br>

Here, $\beta$ does not appear in the equation, as no control measures (such as a reforestation policy) have been put in place.
Finally, assuming that $\lambda_t$ is identical for both countries (due to their similar characteristics and the validity of the parallel trend assumption), the diff-in-diff is then defined as follows:

<br>

\begin{equation}
[E[Y_{it} | \text{Treated}_i, \text{Post}_t] - E[Y_{it} | \text{Treated}_i, \text{Pre}_t]] - [E[Y_{it} | \text{Control}_i, \text{Post}_t] - E[Y_{it} | \text{Control}_i, \text{Pre}_t]] = (\lambda_t + \beta) - \lambda_t = \beta
\end{equation}

<br>

Where $\beta$ represents the effect of Ghana's reforestation policy on its forests.

The underlying hypothesis suggests that by reducing land expansion while increasing agricultural land, the relationship could remain unchanged, potentially merely reversing the trend.

However, this could be explained by several factors, as we saw with the OLS regression. With a triple interaction coefficient (Treatment x Post2010 x Agricultural) of 0.07% (p = 0.739), this interaction is not statistically significant for the model with control variables and time fixed effects (see Appendix 7). This implies that the policy does not significantly modify the relationship between agricultural land and forest cover. This result was in fact expected and does not reject our hypothesis. The model therefore has no significant statistical evidence that the reform has had a direct impact on forest cover or on the relationship. This does not necessarily mean that the reform had no effect, but that the model does not capture it.

## 5. Discussion on causality

It is important to keep a critical eye on our method in order to be able to identify whether the effects obtained are causal or not.

### 5.1 Reverse causality

To begin with, we suspect an endogeneity bias in our study, more precisely an inverse causality between the independent variable (agricultural land) and the dependent variable (forest cover). It is plausible that deforestation influences agriculture. For example, a forest destroyed by a natural event could lead to the use of this land for agriculture.

A potential solution to this problem would be to use an instrumental variable Z, which would only affect agricultural areas and not forest cover. However, our model uses data from a large sample of countries, making it difficult to find a relevant instrumental variable that affects the whole of our sample. It would therefore be preferable to focus on a smaller sample, or even a single country, where an agricultural subsidy policy could be used as an instrumental variable.

### 5.2 Omitted-variable-bias

Omitted variable bias remains a potential problem in our model. For example, land management policies vary across countries and influence agriculture and forestry (Han & Hee Go, 2017) <a href="https://doi.org/10.1016/j.landusepol.2018.11.035" target="_blank">[28]</a>, which our model does not take into account because it is complex to measure and compare these policies internationally.

In addition, technological differences in agriculture between countries can bias our model. Although GDP per capita can reflect differences in wealth and, by extension, the level of technology, a more specific variable targeting agricultural productivity would be more appropriate. For example, developing countries use agricultural expansion methods (shifting agriculture and slash and burn method (Curtis et al., 2018) <a href="https://doi.org/10.1126/science.aau3445" target="_blank">[29]</a>) that have a greater impact on forests, unlike developed countries, which intensify production on existing land.


### 5.3 Multicollinearity

Due to the large number of variables considered in the model, our model may potentially suffer from bias due to multicollinearity problems, with FIVs exceeding 10 (see Appendix 3). The choice of variables is therefore very important and we have taken care of this with a correlation analysis (see Appendix 8).

## 6. Conclusion

Although our model has certain weaknesses, we maintain that the negative trend it indicates, namely that the increase in agricultural land is reducing forest area, remains relevant. Deforestation continues to be a reality in most regions of the world, and only a few countries have managed to slow or even reverse this loss of forest. As shown by several studies (Curtis et al., 2018 <a href="https://doi.org/10.1126/science.aau3445" target="_blank">[30]</a>; Geist & Lambin, 2002 <a href="https://academic.oup.com/bioscience/article/52/2/143/341135" target="_blank">[31]</a>; DeFries et al., 2010 <a href="https://www.researchgate.net/publication/248828436_Deforestation_Drive_by_Urban_Population_Growth_and_Agricultural_Trade_in_the_Twenty-First_Century" target="_blank">[32]</a>), the main causes of deforestation include urbanisation, agriculture, forest fires and logging. The results therefore support our hypothesis that the expansion of agricultural land has a negative impact on forest areas. Finally, the model does not clearly identify any significant positive impact of the reforestation policy initiated in Ghana in 2010 on the extent of the country's forests. These results should be treated with caution and require further research.

Nevertheless, we believe it is crucial to look into this issue and identify solutions, following the example of the reforestation policy adopted by Ghana. The world's population is growing steadily, particularly in developing countries such as those in Sub-Saharan Africa. As a result, demand for agri-food products continues to rise. This increase in demand could be met either by intensifying agriculture, leading to improved productivity, or by expanding agricultural land into other areas, such as forests (Zabel et al., 2019) <a href="https://doi.org/10.1038/s41467-019-10775-z" target="_blank">[33]</a>. Thus, we believe there is an urgent need to intervene in developing countries experiencing rapid deforestation, high population growth, and are home to rich biodiversity due to their geographical location (South America, Sub-Saharan Africa, South Asia, Southeast Asia).

### Appendix 6

```python
import statsmodels.api as sm
import statsmodels.formula.api as smf

# indicator variables for pre- posttreatment
dfi['treated'] = (dfi['Iso'] == 'GHA').astype(int)
dfi['post'] = (dfi['Year'] >= 2010).astype(int)

# interaction
dfi['treated_post'] = dfi['treated'] * dfi['post']

df_did = dfi[dfi['Iso'].isin(['GHA', 'CIV'])]

# DiD model
formula = 'Q("Forest Cover") ~ treated + post + treated_post'
model = smf.ols(formula, data=df_did)
results = model.fit()

print(results.summary())
```

```python
from statsmodels.formula.api import ols

# indicator variables for pre- posttreatment
did_data['Treatment'] = np.where(did_data['Iso'] == 'GHA', 1, 0)
did_data['Post2010'] = np.where(did_data['Year'] > 2010, 1, 0)

# Interactions
did_data['TreatmentXPost2010'] = did_data['Treatment'] * did_data['Post2010']
did_data['TreatmentXPost2010XAgri'] = did_data['Treatment'] * did_data['Post2010'] * did_data['Agricultural Land']

# Linear regression with triple interaction
formula = ('Q("Forest Cover") ~ Treatment + Post2010 + Q("Agricultural Land") + '
           'TreatmentXPost2010 + TreatmentXPost2010XAgri + Q("GDP per capita") + '
           'Q("Log Density population") + Q("Latitude") + Q("Log Tree Cover Loss from Fires") + C(Year)')
model = ols(formula, data=did_data).fit()

print(model.summary())
```
